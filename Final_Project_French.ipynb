Si vous utilisez Python pour cet exercice, veuillez inclure le code dans les cellules ci-dessous. Si vous utilisez un autre outil, mettez  les captures d'écran de votre travail.

```python
# renommer la colonne
new_df.rename(columns={'availability 365':'days_booked'})
```

```python
# mettre tous les noms de colonnes en minuscules et remplacer les espaces par un trait de soulignement "_"
new_df.columns = [col.lower().replace(' ', '_') for col in new_df.columns]
new_df.head()
```

```python
# supprimez le signe dollar et la virgule des colonnes price et service_fee. Si nécessaire, convertir les 2 colonnes dans le type de données appropriées

import numpy as np
def bon_prix (ancien_prix):
   
 
    if(pd.isna(ancien_prix) or (ancien_prix=='nan') ):
        return np.NaN

    else:
        prix1=ancien_prix.replace(',', '')
        prix2=prix1.replace(' ','')
        prix3=prix2.replace('$','')
        nouveaux_prix=(prix3.replace('.',''))
       

        return float(nouveaux_prix)
```

```python
df['price']=df['price'].apply(lambda x: bon_prix(x))
```

```python
df['service fee']=df['service fee'].apply(lambda x: bon_prix(x))
```
